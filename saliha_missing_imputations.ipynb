```python
#getting rows where county is null from the dataset
df_fire_null_county = df_fire.loc[pd.isna(df_fire['COUNTY'])]
df_fire_null_county.shape
```

```python
#those with digits records
df_not_null = df_fire.loc[pd.notnull(df_fire['COUNTY'])]
df_digit = df_not_null.loc[df_not_null['COUNTY'].str.isdigit()]
df_digit
```

## Imputing COUNTY using Geopy package

```python
#using latitude and longitude to find the county.
geolocator = Nominatim(user_agent="saliha_missingimputation")
list_county=[] # alternative storage of county
def location(row):
    location = geolocator.reverse(str(row.LATITUDE)+', '+str(row.LONGITUDE),timeout=None)
    try:
        list_county.append(location.raw['address']['county'])
        #print(location.raw['address']['county'])
        return location.raw['address']['county']
    except:
        list_county.append("error")
        #print("eror")
        return 'error'
        
#df_fire_null_county['COUNTY']=df_fire_null_county.apply(lambda x: location(x) ,axis=1)
```
