```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
class DecTreeModel:
    def run_tree(X,y,split):
        from sklearn.model_selection import train_test_split 
        X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=split, random_state=10)
        from sklearn import tree
        tree_model = tree.DecisionTreeClassifier(random_state=1, max_depth=30,min_samples_split=100, min_samples_leaf=10)
        tree_model.fit(X_train,y_train)
        y_pred = tree_model.predict(X_test)
        from sklearn.metrics import classification_report,accuracy_score  
#         print(confusion_matrix(y_test, y_pred))
#         print("Classification Report for Decision Tree")
#         print(classification_report(y_test,y_pred))  
        print("Accuracy score for Decision Tree:",accuracy_score(y_test,y_pred))
        print("Depth of the Original Tree :" + str(tree_model.get_depth()))
        print("No of leaves in the Original Tree :" + str(tree_model.get_n_leaves()))
        
        return
```

```python
DecTreeModel.run_tree(X,y,.3)
```
